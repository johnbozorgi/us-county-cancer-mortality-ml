# Notebook 4 — Geospatial Visualization

```python
import pandas as pd
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

## 1. Load Data

```python
df = pd.read_csv('../data/processed_merged.csv')
df['FIPS'] = df['FIPS'].astype(str).str.zfill(5)
print('Counties to map:', len(df))
```

## 2. Choropleth — Mortality Rate

```python
fig = px.choropleth(
    df,
    geojson='https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json',
    locations='FIPS',
    color='rate',
    color_continuous_scale='Reds',
    scope='usa',
    labels={'rate': 'Cancer Mortality Rate'},
    title='Cancer Mortality Rate by US County (Age-Adjusted, 2019–2023)',
    hover_data={'Area': True, 'rate': ':.1f', 'FIPS': False}
)
fig.update_layout(margin={'r':0,'t':50,'l':0,'b':0})
fig.write_html('../visualizations/cancer_mortality_map.html')
fig.show()
print('Saved: cancer_mortality_map.html')
```

## 3. Scatter: Mortality vs Incidence (Equity Gap)

```python
fig2 = px.scatter(
    df, x='IncidenceRate', y='rate',
    hover_name='Area',
    labels={'IncidenceRate': 'Incidence Rate', 'rate': 'Mortality Rate'},
    title='Mortality vs Incidence Rate — Counties above trend line = equity gap',
    opacity=0.5, trendline='ols'
)
fig2.write_html('../visualizations/mortality_vs_incidence_scatter.html')
fig2.show()
print('Saved: mortality_vs_incidence_scatter.html')
```
